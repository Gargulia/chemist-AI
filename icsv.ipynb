```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('sales_with_goods.csv')
df = df.drop(columns=['tradeMark'])
df = df.drop(columns=['iaCourseOfTreatment'])
df = df.drop(columns=['productFullType'])
df = df.drop(columns=['diseases'])
df = df.drop(columns=['storageCode'])
df = df.drop(columns=['discountCardItem'])
df = df.drop(columns=['discount'])
df = df.drop(columns=['isCashless'])
df = df.drop(columns=['storageId'])
df = df.drop(columns=['ID'])
df = df.drop(columns=['producerItem'])
df = df.drop(columns=['codeEAN'])
df = df.drop(columns=['actualNameItem'])

df.describe()
```

```python
df.goodsGroupItem = df.goodsGroupItem.astype('category')
df.storedCountry = df.storedCountry.astype('category')
df.storedManufacturer = df.storedManufacturer.astype('category')
df.generic = df.generic.astype('category')
df.customerItem = df.customerItem.astype('category')
df.productId = df.productId.astype('category')
df.documentNumber = df.documentNumber.astype('category')
df.analogItem = df.analogItem.astype('category')
df.isGNVLS = df.isGNVLS.astype('category')
df.astraGoodsItem = df.astraGoodsItem.astype('category')

df.describe(include='category')
```

```python
sns.set(context="paper", font="monospace")


corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))

sns.heatmap(corrmat, vmax=.8, square=True)
```
